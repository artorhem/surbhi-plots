### Zoned HMLS for 4K writes. (lsdm/zoned/200GB/qd1/bs4KB/round2/lat_log_lat.1.log)

```python
df1 = pd.read_csv('lsdm/zoned/200GB/qd1/bs4KB/round2/lat_log_lat.1.log', sep=',', names=['op1', 'latency', 'op2', 'write_size', 'op3'])
# Convert nanoseconds to microseconds for the 'time' column
df1['latency'] = df1['latency'] / 1000  # Convert nanoseconds to microseconds

# filtered_df = df1[df1['latency'] > 40000]
# filtered_df.describe()
# print(filtered_df)

# Calculate cumulative GBs of data written
df1['cumulative_gb'] = df1['write_size'].cumsum() / (1024**3)

#Create the line plots
# plt.plot(df1.index, df1['latency'], linestyle='-')
plt.plot(df1['cumulative_gb'], df1['latency'], linestyle='-')

# Add labels and title
plt.xlabel('Cumulative GBs of Data Written')
# plt.xticks(np.arange(0, 16, 1))
plt.ylabel('Latency (microseconds)')
# plt.yticks(np.arange(0, 100001, 10000))
plt.title('HM-LS latency of random writes under zoned 90:10 workload\nbs=4K, qd=1; Total writes: 200GB')

# Show the plot
plt.grid(True)  # Optionally, add grid lines
plt.savefig('fig/lsdm_zoned_200GB_qd1_bs4KB_round2_lat_write.png')
plt.show()
```
