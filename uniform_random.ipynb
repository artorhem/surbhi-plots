```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import seaborn as sns

names_bw=['time', 'bw', 'op1', 'size', 'op2' ]
names_latency=['op1', 'latency', 'op2', 'write_size', 'op3']
```

### Uniform Random on HM-LS

```python
df_latency = pd.read_csv('remainder/uniform/lsdm/uniform/fullDisk/qd1/bs1MB/lat_log_lat.1.log', names=names_latency)
#process the bw data
df_latency['latency'] = df_latency['latency'] / (1000 *1000)  # Convert nanoseconds to milliseconds
# Calculate cumulative GBs of data written
df_latency['cumulative_gb'] = df_latency['write_size'].cumsum() / (1024**3)
```

```python
#Create the line plots
plt.figure(figsize=(8, 6), dpi=80)
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42


fig, ax = plt.subplots()
ax.plot(df_latency['cumulative_gb'], df_latency['latency'], linestyle='-')

# Add labels and title
plt.xlabel('Cumulative GBs Written', fontsize=16)

# xticks = np.arange(0, df_latency['cumulative_gb'].to_list()[-1] + 10 , 10)

# xtick_labels = [str(x) for x in xticks]
# label_formats = [r'$\mathrm{%s}$' % label if i != len(xtick_labels) else r'$\mathbf{\color{red}{%s}}$' % label for i, label in enumerate(xtick_labels)]

# plt.xticks(xticks, label_formats, fontsize=14)
plt.xticks(fontsize=14)
plt.ylabel('Latency (ms)', fontsize=16)


# plt.yticks(np.arange(0, 25000, 2500))
plt.yticks(fontsize=14)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
# plt.title('Latency of HMLS with uniform random\nIO Pattern(200GB, bs=1MB)', fontsize=16)
#set figure size
average_latency = df_latency['latency'].mean()
average_latency_str = str(round(average_latency, 2)) + 'ms'
plt.axhline(y=average_latency, color='r', linestyle='-', label='Avg Latency: ' + average_latency_str)
labels = plt.gca().get_xticklabels()
label_to_change = labels[-1].set_color('red')
plt.legend(loc='upper left', fontsize=14)

# Show the plot
plt.grid(True)  # Optionally, add grid lines
plt.savefig('fig/final/uniform_random/hmls_uniform_fulldisk_1mb_latency.pdf', bbox_inches='tight')
plt.show()
```
